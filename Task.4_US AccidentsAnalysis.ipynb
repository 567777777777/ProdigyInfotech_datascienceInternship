```python
#Data imputation
df_imp = df.copy()
df_imp = df_imp.dropna(subset="City")
df_imp["Temperature(F)"].fillna(df_imp["Temperature(F)"].mean(), inplace=True)
proportions_weather = df_imp["Weather_Condition"].value_counts(normalize=True)

def impute_weather_condition(row):
    if pd.isna(row["Weather_Condition"]):
        return np.random.choice(proportions_weather.index, p=proportions_weather)
    else:
        return row["Weather_Condition"]

df_imp["Weather_Condition"] = df_imp.apply(impute_weather_condition, axis=1)
```

```python
#Analysis and Visualization

df_col = df_imp.copy()
df_col.head()
```

```python
df_col["Start_Time"][0]
'2016-02-08 05:46:00'
df_col["Start_Time"] = pd.to_datetime(df_col["Start_Time"], format="mixed")
```

```python
sns.distplot(df_col["Start_Time"].dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
fig, ax =plt.subplots(1,2, figsize=(12,6))
weekday_start_time = df_col["Start_Time"][(df_col["Start_Time"].dt.day_of_week != 6) & (df_col["Start_Time"].dt.day_of_week != 5)]
sns.distplot(weekday_start_time.dt.hour, bins=24, kde=False, norm_hist=True, ax=ax[0])
ax[0].set_title('Weekdays')

weekend_start_time = df_col["Start_Time"][(df_col["Start_Time"].dt.day_of_week == 6) | (df_col["Start_Time"].dt.day_of_week == 5)]
sns.distplot(weekend_start_time.dt.hour, bins=24, kde=False, norm_hist=True, ax=ax[1])
ax[1].set_title('Weekend')

plt.subplots_adjust(wspace=0.2)
fig.show()
```
